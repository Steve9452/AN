```python
from sympy import Symbol, Rational
from math import exp, cos, sin
```

```python
x = Symbol('x')
y = Symbol('y')
```

```python
def lagrangePol(n, xs, ys):
    p = 0
    for k in range(0, n+1):
        p += ys[k] * lagrangeL(n, xs, k)
    return p 
```

```python
def lagrangeL(n, xs, k):
    L = 1
    for i in range(0, n+1):
        if (i != k):
            L *=(x - xs[i] )/(xs[k] - xs[i])
    return L
```

```python
xs = [0, -1, 1]
ys = [1, 2, 2]

pol = lagrangePol(2, xs, ys)
pol.expand()
```

```python
pol.subs(x, 1)
```
